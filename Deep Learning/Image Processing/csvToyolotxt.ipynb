```python
import csv
import os
import glob
import pandas as pd
```

```python
df=pd.read_csv('./aug_gray_train_10.csv')

print(df.dtypes)
print(df)
```

```python
print(len(df))
```

```python
for i in range(0,len(df)):
    
    filename=df.iloc[i]['filename']
    classes_names=df.iloc[i]['class']
    xmin=df.iloc[i]['xmin']
    ymin=df.iloc[i]['ymin']
    xmax=df.iloc[i]['xmax']
    ymax=df.iloc[i]['ymax']
    
    
    txtfilename='./gray_train2/aug_gray_train_10/'+filename[:-4]+'.txt'
    
    print(txtfilename)
    
    f=open(txtfilename,'w')
    if(classes_names=='wardrobe'):
        f.write('1')
    else:
        f.write('0')
        
    dw=1./416
    dh=1./416
    x=(xmin+xmax)/2.0
    y=(ymin+ymax)/2.0
    w=xmax-xmin
    h=ymax-ymin
    x=x*dw
    w=w*dw
    y=y*dh
    h=h*dh
    
    x=str(x)
    y=str(y)
    w=str(w)
    h=str(h)
    
    f.write(' '+x+' '+y+' '+w+' '+h)
    
    f.close()
        
    
```
